```python
import pickle
import numpy as np  # Needed for handling input arrays
import cv2  # Required if preprocessing images
```

```python
# Load the model from the pickle file
with open('brain_tumor_model.pkl', 'rb') as file:
    loaded_model = pickle.load(file)

print("Model loaded successfully!")
```

```python
# Load an example image
img = cv2.imread('Te-no_0042.jpg')
img = cv2.resize(img, (150, 150))  # Resize to match model's input size
img_array = np.array(img)  # Convert to NumPy array
img_array.shape
```

```python
img_array = img_array.reshape(1, 150, 150, 3)  # Reshape to include batch dimension
img_array.shape
```

```python
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image
img = image.load_img('Te-no_0042')
plt.imshow(img,interpolation='nearest')
plt.show()
```

```python
# Make a prediction
prediction = loaded_model.predict(img_array)

indices = prediction.argmax()

if(indices==2):
    print("No Tumor Detected")
else:
    print("Tumor Detected")
```
